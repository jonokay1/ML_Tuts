### Recommendation Engine Notes

Data used for this recomendation engine uses the names of the FATHERS of Computer Science and Artificial Intelligence
* In this tut, _collaborative filtering_ is used
* The idea behind collaborative filtering is that if person A likes 1,2,3 and B likes 2, 3, 4 then they have similar interests. A should love 4 and B should love 1

* The other form of recommendation Engine not used here is _Content Based Algorithms_.
* Content based algorithms assume if you like a specific kind of item... e.g movies with a particular genre, then you will like another movie in that genre.

* The data is in a dictionary that contains the items (cs system books) and a relative score the scientists attached to them
* We seek to determine which books the different scientists are most likely to read based on how they rated the ones they had read
* First we determine how similar people are based on thei ranking
* We then find an adequate similarity score using _Euclidean Distance_ and _Pearson Correlation Coefficient_

#### Euclidean Distance Score
* The Euclidean distance between two points is the length of the line segments connecting them.
* The Euclidean distance or Euclidean metric is the "ordinary" straight-line distance between two points in Euclidean space. 

* Two people belong to a certain preference space if and only if they have ranked the two items that define that preference space. 
* The closer two people are in a preference space, the closer their similarity

#### Pearson Correlation Coefficient
* The Pearson correlation Coefficient is a measure of the linear dependance or correlation between two variables X and Y
We use the common datapoints (common fields/books) between 2 people so as to determine the similarity between users

* Plotting these would give us a positive or negative slope indicating the correlation

#### Recommending
* Using the methods above, we can then make a calculated guess on what the Scientist would find interesting
* These methods help us find the closest person to a specific person

* For normalisation, all the ratings on a book by the different readers can be taken into account
* All the ratings by the other users are multiplied by the rating of a specific person. They are then divided by the weight of the sum of the people that have similar preferences.
